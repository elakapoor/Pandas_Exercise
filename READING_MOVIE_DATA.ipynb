```python
df['duration'] = df['duration'].str.slice(0 , -5)
```

```python
df1 = df['duration'].str.split('.', expand=True).astype(int)
df['duration'] = df1 * 60
```

```python
movie_crew_df = pd.read_csv('crew_data.tsv', sep='\t')
```

```python
movie_crew_df
```

```python
movie_crew_df.rename(columns={ 
        'directors' : 'director_ids',
        'tconst' : 'ID'
    }, inplace=True)


movie_crew_df.columns
```

```python
movie_crew_df
```

```python
del movie_crew_df['writers']
```

```python
movie_crew_df
```

```python
df_merge_col = pd.merge(df, movie_crew_df, on = 'ID')
df_merge_col
```

```python
df_merge_col.info()
```

```python
df_merge_col['director_ids_Count'] = df_merge_col['director_ids'].str.len()
df_merge_col
```

```python
df_merge = df_merge_col.loc[df_merge_col['director_ids_Count'] == 9] 
df_merge
```

```python
df_merge.rename(columns={ 
        'director_ids' : 'director_id',
    }, inplace=True)


df_merge.columns
```
